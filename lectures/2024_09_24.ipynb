```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Defining constants
# File
FILENAME: str = "iris.csv"
ROOT_DIR: str = os.path.join(os.getcwd(), os.pardir)
RESOURCES_DIR: str = os.path.join(ROOT_DIR, "resources")

# Plotting
FIG_SIZE: tuple[int, int] = (6, 6)
```

```python
# Defining parameters
SEED: int = 12
TEST_SIZE: float = 0.25
```

```python
# Defining column names
sepal_length: str = "sepal length(cm)"
sepal_width: str = "sepal width(cm)"
petal_length: str = "Petal length(cm)"
petal_width: str = "Petal Width(cm)"
species: str = "Species"
```

```python
# Defining functions
def read_file(filename: str) -> pd.DataFrame:
    filepath: str = os.path.join(RESOURCES_DIR, filename)
    return pd.read_csv(filepath)


def print_shapes(*datasets: pd.DataFrame | np.ndarray) -> None:
    shapes = [str(dataset.shape) for dataset in datasets]
    print(", ".join(shapes))


def create_train_and_test_set(X: pd.DataFrame, y: pd.DataFrame) -> tuple[
    np.ndarray, np.ndarray, np.ndarray, np.ndarray
]:
    X_set: np.ndarray = np.array(X)
    y_set: np.ndarray = np.array(y)

    train_test_splits: tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray] = train_test_split(
        X_set,
        y_set,
        test_size=TEST_SIZE,
        random_state=SEED
    )

    return train_test_splits


def plot_scatter(x: np.ndarray, y: np.ndarray, title: str) -> None:
    sns.scatterplot(x=x, y=y, label=title)
    plt.show()


def find_outliers(dataframe: pd.DataFrame) -> np.ndarray:
    index_array: list[int] = []
    columns: np.ndarray = dataframe.columns.to_numpy()

    for column in columns:
        Q1: float = dataframe[column].quantile(0.25)
        Q3: float = dataframe[column].quantile(0.75)
        IQR: float = Q3 - Q1

        lower_bound: float = Q1 - 1.5 * IQR
        upper_bound: float = Q3 + 1.5 * IQR

        indexes: np.ndarray = dataframe[
            (dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)
            ].index.to_numpy()

        for index in indexes:
            index_array.append(index)

    return np.array(index_array)
```

#### Simple preprocessing

```python
df: pd.DataFrame = read_file(filename=FILENAME)
df: pd.DataFrame = df.reset_index(drop=True)
X: pd.DataFrame = df[[sepal_length, sepal_width, petal_length, petal_width]]
y: pd.DataFrame = df[[sepal_length]]
```

```python
correlation_matrix: pd.DataFrame = X.corr()
plt.figure(figsize=FIG_SIZE)
sns.heatmap(correlation_matrix, annot=True)
```

```python
variables: list[str] = [sepal_length, sepal_width, petal_length, petal_width]

plt.figure(figsize=(12, 8))
for i, var in enumerate(variables, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df[var])
    plt.title(f'Box plot of {var}')

plt.tight_layout()
plt.show()
```

```python
plot_scatter(df[sepal_width], df[sepal_length], "Sepal width vs Sepal length")
plot_scatter(df[petal_width], df[sepal_length], "Petal width vs Sepal length")
plot_scatter(df[petal_length], df[sepal_length], "Petal  length vs Sepal length")
```

#### Split into train and test sets

```python
features: list[str] = [sepal_width, petal_length]
output: list[str] = [sepal_length]
X: pd.DataFrame = df[features]
splits: tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray] = create_train_and_test_set(X, y)

X_train: np.ndarray = splits[0]
X_test: np.ndarray = splits[1]
y_train: np.ndarray = splits[2]
y_test: np.ndarray = splits[3]

X_train: pd.DataFrame = pd.DataFrame(X_train, columns=features)
X_test: pd.DataFrame = pd.DataFrame(X_test, columns=features)
y_train: pd.DataFrame = pd.DataFrame(y_train, columns=output)
y_test: pd.DataFrame = pd.DataFrame(y_test, columns=output)

print_shapes(X_train, y_train, X_test, y_test)
```

#### Preprocessing
From the EDA we observe the following:
- The box-plots show that there are four outliers in the sepal width column.
- The correlation matrix shows petal widht and petal length are highly correlated with sepal length

```python
outlier_indexes: np.ndarray = find_outliers(X_train)
X_train: pd.DataFrame = X_train.drop(index=outlier_indexes)
y_train: pd.DataFrame = y_train.drop(index=outlier_indexes)
print_shapes(X_train, y_train)
```

#### Train model

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
```

#### Predict and evaluate model

```python
predictions: np.ndarray = linear_regression.predict(X_test)
mae: float = metrics.mean_absolute_error(y_test, predictions)
mse: float = metrics.mean_squared_error(y_test, predictions)
rmse: float = np.sqrt(mse)
r2_score: float = metrics.r2_score(y_test, predictions)
```

```python
print(
    f"The predictions gave the following results:\n" +
    f"- R-squared: \t{r2_score}\n" +
    f"- MAE:       \t{mae}\n" +
    f"- MSE:       \t{mse}\n" +
    f"- RMSE:      \t{rmse}"
)
```
